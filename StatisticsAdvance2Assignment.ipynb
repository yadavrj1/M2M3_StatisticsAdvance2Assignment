Question 1: What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) is a statement that there is no effect, no difference, or no relationship in the population. It represents the default or status-quo assumption that any observed effect in your sample data is due to random chance.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true (Type I error). It represents the threshold for deciding whether the observed results are statistically significant. Commonly set at 0.05 (5%), it means there is a 5% risk of concluding there is an effect when none exists. If the p-value from the test is less than α, we reject the null hypothesis and accept that there is enough evidence for the alternative hypothesis. Thus, α helps control false positives and ensures decisions are made with an acceptable level of risk.

Question 4: What are Type I and Type II errors? Give examples of each.
  - 1. Type I Error (False Positive)
- Definition: Rejecting the null hypothesis (H₀) when it is actually true.
- Meaning: Concluding there is an effect or difference when none exists.
- Example: A medical test indicates a patient has a disease when they actually do not.

  - 2. Type II Error (False Negative)
- Definition: Failing to reject the null hypothesis when it is actually false.
- Meaning: Missing a real effect or difference.
- Example: A medical test fails to detect a disease in a patient who actually has it.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
  - Z-test Checking if average height of students equals national average when population variance is known. and T-test Checking if average height of students equals national average when population variance is unknown.
  - When to use:
  - Z-test: Use when population variance is known.
  - T-test: Use when population variance is unknown.

```python
''' Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.) '''

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Std Dev: {sample_std:.3f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the Null Hypothesis (H₀).")
else:
    print("Fail to Reject the Null Hypothesis (H₀).")
```

```python
''' Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.) '''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence = 0.95
z_value = stats.norm.ppf(1 - (1 - confidence)/2)
margin_of_error = z_value * (std / np.sqrt(n))
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"Sample Standard Deviation: {std:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

plt.figure(figsize=(8,5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.axvline(lower_bound, color='green', linestyle='--', label=f"95% CI Lower = {lower_bound:.2f}")
plt.axvline(upper_bound, color='green', linestyle='--', label=f"95% CI Upper = {upper_bound:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
''' Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.) '''

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):

    mean = np.mean(data)
    std = np.std(data, ddof=1)

    z_scores = (data - mean) / std

    print(f"Mean of data: {mean:.2f}")
    print(f"Standard Deviation: {std:.2f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")

    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=15, color='lightblue', edgecolor='black')
    plt.title('Histogram of Standardized Data (Z-scores)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

data = np.random.normal(loc=100, scale=15, size=200)
z_scores = calculate_z_scores(data)
```
